# Haberman's Survival Data Analysis 

### Description:

- The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.

### Relevant Information: 

- The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.

- Number of Instances: 306

- Number of Attributes: 4 (including the class attribute)

- Attribute Information:

    - Age of patient at time of operation (numerical)
    - Patient's year of operation (year - 1900, numerical)
    - Number of positive axillary nodes detected (numerical)
    - Survival status (class attribute) 1 = the patient survived 5 years or longer 2 = the patient died within 5 year

- Missing Attribute Values: None

### Histogram, CDF, PDF :

```python
survived = hbman.loc[hbman['status'] == 1 ];
ntsurvived = hbman.loc[hbman['status'] == 2 ];

plt.plot(survived["year"] , np.zeros_like(survived), 'o')
plt.plot(ntsurvived["year"], np.zeros_like(ntsurvived), 'o')
plt.show()
plt.plot(survived["age"] , np.zeros_like(survived), 'o')
plt.plot(ntsurvived["age"], np.zeros_like(ntsurvived), 'o')
plt.show()
plt.plot(survived["nodes"] , np.zeros_like(survived), 'o')
plt.plot(ntsurvived["nodes"], np.zeros_like(ntsurvived), 'o')
plt.show()
```

```python
sns.FacetGrid(hbman, hue="status", size=5) \
   .map(sns.distplot, "year") \
   .add_legend();
plt.show();

sns.FacetGrid(hbman, hue="status", size=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();

sns.FacetGrid(hbman, hue="status", size=5) \
   .map(sns.distplot, "nodes") \
   .add_legend();
plt.show();
```

  ### CDF AND PDF:

```python
# probability density function and cummalative distribution function 

counts, bin_edges = np.histogram(survived['age'], bins=10, density = True)
print (counts) ;
print (bin_edges) ;
pdf = counts/(sum(counts))
print ("Histogram of Survived - Age")
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(survived['year'], bins=10, density = True)
print (counts) ;
print (bin_edges) ;
pdf = counts/(sum(counts))
print ("Histogram of Survived - Year")
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(survived['nodes'], bins=10, density = True)
print (counts) ;
print (bin_edges) ;
pdf = counts/(sum(counts))
print ("Histogram of Survived - Nodes")
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)

plt.show() 

counts, bin_edges = np.histogram(ntsurvived['age'], bins=10, density = True)
print (counts) ;
print (bin_edges) ;
pdf = counts/(sum(counts))
print ("Histogram of Not survived - age")
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(ntsurvived['year'], bins=10, density = True)
print (counts) ;
print (bin_edges) ;
pdf = counts/(sum(counts))
print ("Histogram of Not survived - Year")
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(survived['nodes'], bins=10, density = True)
print (counts) ;
print (bin_edges) ;
pdf = counts/(sum(counts))
print ("Histogram of Not survived - Nodes")
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.show()
```

### Mean, Variance and Std-dev

```python
print("Means:")
print ("Mean age Survived and Not-survived")
print(np.mean(survived["age"]))
print(np.mean(ntsurvived["age"]))
print ("Mean Year Survived and Not-Survived")
print(np.mean(survived["year"]))
print(np.mean(ntsurvived["year"]))
print ("Mean node Survived and Not-Survived")
print(np.mean(survived["nodes"]))
print(np.mean(ntsurvived["nodes"]))

print("\nStd-dev:");
print ("Std-Deviation age Survived and Not-survived")
print(np.std(survived["age"]))
print(np.std(ntsurvived["age"]))
print ("Std-Deviation year Survived and Not-survived")
print(np.std(survived["year"]))
print(np.std(ntsurvived["year"]))
print ("Std-Deviation nodes Survived and Not-survived")
print(np.std(survived["nodes"]))
print(np.std(ntsurvived["nodes"]))
```

### Median, Percentile, Quantile, IQR, MAD:

```python
print("\nMedians:")
print ("median of survived and not survived - Age" )
print(np.median(survived["age"]))
print(np.median(ntsurvived["age"]))
print ("median of survived and not survived - Year" )
print(np.median(survived["year"]))
print(np.median(ntsurvived["year"]))
print ("median of survived and not survived - Nodes" )
print(np.median(survived["nodes"]))
print(np.median(ntsurvived["nodes"]))


print("\nQuantiles:")
print ("Percentile of survived and not survived - Age" )
print(np.percentile(survived["age"],np.arange(0, 100, 25)))
print(np.percentile(ntsurvived["age"],np.arange(0, 100, 25)))
print ("Percentile of survived and not survived - Year" )
print(np.percentile(survived["year"],np.arange(0, 100, 25)))
print(np.percentile(ntsurvived["year"],np.arange(0, 100, 25)))
print ("Percentile of survived and not survived - Nodes" )
print(np.percentile(survived["nodes"],np.arange(0, 100, 25)))
print(np.percentile(ntsurvived["nodes"],np.arange(0, 100, 25)))

print("\n90th Percentiles:")
print ("90th Percentile of survived and not survived - Age" )
print(np.percentile(survived["age"],90))
print(np.percentile(ntsurvived["age"],90))
print ("90th Percentile of survived and not survived - Year" )
print(np.percentile(survived["year"],90))
print(np.percentile(ntsurvived["year"],90))
print ("90th Percentile of survived and not survived - Nodes" )
print(np.percentile(survived["nodes"],90))
print(np.percentile(ntsurvived["nodes"],90))

from statsmodels import robust
print ("\nMedian Absolute Deviation")
print ("Mean std dev. of survived and not survived - Age" )
print(robust.mad(survived["age"]))
print(robust.mad(ntsurvived["age"]))
print ("Mean std dev. of survived and not survived - Year" )
print(robust.mad(survived["year"]))
print(robust.mad(ntsurvived["year"]))                    
print ("Mean std dev. of survived and not survived - Nodes" )
print(robust.mad(survived["nodes"]))
print(robust.mad(ntsurvived["nodes"]))      
```
